```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'Global_AI_Content_Impact_Dataset.csv')
#pd.set_option('display.max_columns', None)
df.head()
```

```python
df.isnull().sum()
```

```python
# Categorical columns
categorical_col = df.select_dtypes(include=['object', 'category'])

# Numerical columns
numerical_col = df.select_dtypes(include=['number'])
print('categorical columns==> ',categorical_col.columns)
print('numerical columns==> ',numerical_col.columns)
```

```python

m=np.triu(numerical_col.corr())
sns.heatmap(numerical_col.corr(), annot=True,mask=m, cmap="coolwarm")
```

```python
sns.histplot(df['AI Adoption Rate (%)'], bins=20,kde=10)
plt.show()
```

```python
sns.histplot(df['Job Loss Due to AI (%)'], bins=20,kde=10)
plt.show()
```

```python
c=df.groupby('Industry')['Job Loss Due to AI (%)'].sum()
plt.pie(c,labels=c.index, autopct='%0.1f%%')
plt.title("")
plt.axis('equal')
plt.show()
```
